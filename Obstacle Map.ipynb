```python
def fillOneNeighbor(p, q, n, m, obsmap, curr):
    #if it's outside the box
    if p < 0 or q < 0 or p > n-1 or q > m-1:
        #go to next neighbor
        return
    #if the value is -1
    elif obsmap[p,q] == -1:
        #go to next neighbor
        return
    #if it's the start point
    elif p==0 and q==0:
        #label it with curr+1
        obsmap[p,q] = curr+1
        #stop execution
        return
    #if it's zero
    elif obsmap[p,q] == 0:
        #label it with curr+1
        obsmap[p,q] = curr+1
        return

def fill(obsmap, n, m):
    #set end to 1
    obsmap[19,19] = 1
    #set current number to 1
    curr = 1
    
    #notFinished = True
    
    #while true
    while curr < n*m:
         #Go through the whole array
         for i in range(n):
             for j in range(m):
                 #if this point is our current number
                 if obsmap[i,j] == curr:
                     #for each neighbor
                     p = i-1
                     q = j
                     fillOneNeighbor(p,q,n,m,obsmap,curr)
                     p = i
                     q = j-1
                     fillOneNeighbor(p,q,n,m,obsmap,curr)
                     p = i+1
                     q = j
                     fillOneNeighbor(p,q,n,m,obsmap,curr)
                     p = i
                     q = j+1
                     fillOneNeighbor(p,q,n,m,obsmap,curr)
         #increment current number
         curr = curr+1
    print(obsmap)
    return obsmap

def descendOneNeighbor(p,q,i,j,n,m,obsmap,startVal, storedPts):
    #if the index is invalid
    if p < 0 or q < 0 or p > n-1 or q > m-1:
        # go to next neighbor
        return i,j,startVal
    #if current value is one less than starting value
    elif obsmap[p,q] == startVal-1:
        #Store that index
        storedPts.append((p,q))
        #Move to that index
        i = p
        j = q
        startVal = startVal-1
        #if current value is 1
        if startVal == 1:
            #print the path index
            print(storedPts)
        return i,j,startVal
    return i,j,startVal

def descend(obsmap,n,m):     
    storedPts = []
    #Start at (0,0)
    i = 0
    j = 0
    #Store the starting value as 47
    startVal = obsmap[i,j]
    #while we haven't reached the end
    while True:
        #for each of the neighbors
        p = i+1
        q = j
        i,j,startVal = descendOneNeighbor(p,q,i,j,n,m,obsmap, startVal, storedPts)
        p = i
        q = j+1
        i,j,startVal = descendOneNeighbor(p,q,i,j,n,m,obsmap, startVal, storedPts)
        p = i-1
        q = j
        i,j,startVal = descendOneNeighbor(p,q,i,j,n,m,obsmap, startVal, storedPts)
        p = i
        q = j-1
        i,j,startVal = descendOneNeighbor(p,q,i,j,n,m,obsmap, startVal, storedPts)
        if i == 19 and j == 19:
            break
        
        

def main():
    getConfig()
    file_dict = getConfig()
    print(file_dict)
    obsmap, n, m = openobsmapFile()

    obsmap = fill(obsmap, n, m)
    descend(obsmap,n,m)
    
main()
```
